```python
#Scikit-Learn
import pandas as pd
from sklearn.datasets import load_iris
data=load_iris()
print(data)
#Target names are features that can be predicted
column_names=data.feature_names
print(column_names)
target_name=data.target_names
print(target_name)
```

```python
df=pd.DataFrame(data.data,columns=column_names)
print(df.to_string())
```

```python
df.columns=['sepal_length','sepal_width','petal_length','petal_width']
print(df.to_string())
```

```python
df.head()
df.shape
df.corr()
df.describe()
```

```python
df.plot.hist(x='sepal_length',y='sepal_width',title='Sepal Length vs Width')
```

```python
df.plot.pie(y='sepal_length',title='Sepal Length')
```

```python
x=df['sepal_length'].values.reshape(-1,1)
y=df['sepal_width'].values.reshape(-1,1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
print(x_train)
print("*********************")
print(x_test)
print("*********************")
print(y_train)
print("*********************")
print(y_test)
```

```python
z=x_train
print(z)
z.shape
k=x_test
print(k)
k.shape
```

```python
from sklearn.linear_model import LinearRegression
mol=LinearRegression()
mol.fit(x_train,y_train)
print(mol.intercept_)
print(mol.coef_)
```

```python
#Predictions
score=mol.predict([[5.8]])
print(score)
```
