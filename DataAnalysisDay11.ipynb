```python
import pandas as pd
heart=pd.read_csv(r"C:\Users\User\Downloads\heart.csv")
print(heart)
```

```python
heart.describe()
```

```python

heart=pd.DataFrame(heart)
duplicates = heart.duplicated(keep=False)


print(heart[duplicates])
```

```python
#Exploratory data analysis
#Variance
print("Variance:\n", heart.var())
```

```python
print("Mean:\n",heart.mean())
print()
print("Mode:\n",heart.mode().iloc[0])

print("Median:\n",heart.median())
print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Heatmap
plt.figure(figsize=(10, 10))
corr_matrix = heart.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap ')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.histplot(heart['age'], kde=True, bins=10)
plt.title('Age Against CP')
plt.xlabel('age')
plt.ylabel('cp')
plt.show()
```
